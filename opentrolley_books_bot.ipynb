```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import datetime
```

```python
class opentrolley_bot:
    
    def __init__(self, url):
        self.url = url
    
    def create_soup(self, sleep = 3):
        time.sleep(sleep)
        s = requests.Session()
        response = s.get(self.url, timeout = 60)
        if response.status_code == 200:
            page = requests.get(self.url).text
            soup = BeautifulSoup(page, 'lxml')
        else:
            print('Not successful response')
        return soup
    
    def extract_info(self, sleep = 3):
        soup = self.create_soup(sleep)
        
        try:
            desc = soup.find('div', {'class':'book-synopsys'}).text.strip()
            whitelist = set('abcdefghijklmnopqrstuvwxyz ABCDEFGHIJKLMNOPQRSTUVWXYZ')
            myStr = desc.text.replace("\n", " ").replace("\r", " ")
            desc_revised = ''.join(filter(whitelist.__contains__, myStr))
            desc_revised = desc_revised.strip()
        except:
            desc_revised = None
        
        try:
            title = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblTitle'}).text.strip()
        except:
            title = None
        
        try:
            ean = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblEAN'}).text.strip()
        except:
            ean = None
        
        try:
            author = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblContributor1'}).text.strip()
        except:
            author = None
        
        try:
            bind = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblBindingDescription'}).text.strip()
        except:
            bind = None
        
        try:
            discounted_price = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblDiscountedPrice'}).text.strip()
        except:
            discounted_price = None
        
        try:
            normal_price = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblNormalPrice'}).text.strip()
        except:
            normal_price = None
        
        try:
            stock = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblStockStatus'}).text.strip()
        except:
            stock = None
        
        try:
            copies = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblstock'}).text.strip()
        except:
            copies = None
        
        try:
            published_date = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblPublicationDate'}).text.strip()
        except:
            published_date = None
        
        try:
            about_author = soup.find('span', {'id': 'ctl00_ContentPlaceHolder1_lblAuthorDescr1'}).text.strip()
        except:
            about_author = None
        
        info_dict = {'EAN': ean,
                    'Title': title,
                    'Author': author,
                    'Normal_Price': normal_price,
                    'Discounted_Price': discounted_price,
                    'Description': desc_revised,
                    'Stock': stock,
                    'Copies': copies,
                    'Bind': bind,
                    'Published_Date': published_date,
                    'About_Author': about_author}
        
        return info_dict
```
