```python
import pandas as pd
import numpy as np
import datetime
import wikipediaapi
```

```python
# define class
class telco_wikipedia:
    
    def __init__(self, title):
        self.title = title
    
    def get_page(self):
        wiki = wikipediaapi.Wikipedia(
        language = 'en',
        extract_format = wikipediaapi.ExtractFormat.WIKI
        )
        page = wiki.page(self.title)
        return page

    def get_page_text(self):
        page = self.get_page()
        text = page.text
        return text
    
    def get_page_sections(self):
        page = self.get_page()
        sections = page.sections
        return sections
    
    def get_history(self):
        sections = self.get_page_sections()
        history_list = []
        for item in self.get_page_sections():
            if item.title == 'History':
                for item2 in item.sections:
                    item2_split = item2.text.split(sep = '\n')
                    for item3 in item2_split:
                        if len(item3) > 1:
                            item3_dict = {'Year': item2.title,
                                         'Activity': item3,
                                         'Datetime': datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
                            history_list.append(item3_dict)
        history_df = pd.DataFrame(history_list)
        file_title = self.title.strip().lower().replace(' ','_').replace('(','').replace(')','')
        filename = f"wikipediaapi_{file_title}_{datetime.datetime.now().strftime('%Y%m%d%H%M%S')}.csv"
        history_df.to_csv(filename, index = False)
        return history_df
```

```python
m1 = telco_wikipedia(title = 'M1 (Singaporean company)')
```

```python
m1.get_history()
```
